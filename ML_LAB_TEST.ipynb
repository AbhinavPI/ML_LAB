```python
import pandas as pd
import numpy as np
new_data = {
    'Temperature': np.random.randint(15, 35, 200),
    'Time_of_Day': np.random.randint(0, 24, 200),
    'Light_Intensity': np.random.randint(100, 900, 200),
    'Power_Consumption': np.random.randint(50, 200, 200)
}
new_df_1 = pd.DataFrame(new_data)
new_df_1
to_csv = new_df_1.to_csv('new_data.csv', index=False)
new_df = pd.read_csv('new_data.csv')
new_df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
new_df = pd.read_csv('new_data.csv')
features = ['Temperature', 'Time_of_Day', 'Light_Intensity']
target = 'Power_Consumption'
X_train, X_test, y_train, y_test = train_test_split(new_df[features], new_df[target], test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

new_conditions = pd.DataFrame({
    'Temperature': [25],
    'Time_of_Day': [10],
    'Light_Intensity': [500]
})

predicted_power = model.predict(new_conditions)
print("Predicted Power Consumption:", predicted_power)
rmse = mean_squared_error(y_test, y_pred)
print(f"Root Mean Squared Error (RMSE): {rmse}")

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)
rmse_dt = mean_squared_error(y_test, y_pred_dt)
print(f"Decision Tree Regressor RMSE: {rmse_dt}")
if rmse < rmse_dt:
    print("Linear Regression performs better.")
else:
    print("Decision Tree Regressor performs better.")
```
